```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_dataframe = pandas.read_csv('train.csv')
train_dataframe.head()
```

```python
for c in train_dataframe.columns[1:]:
    ratio = train_dataframe[c].isnull().sum() / train_dataframe[c].size
    if ratio > 0.0:
        print('%s: %.02e' %(c,ratio))
```

```python
train_dataframe.dropna(axis=1, how='any', inplace=True)
train_dataframe.head()
```

```python
class_labels={}
for c in ('MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope'):
    class_labels[c] = train_dataframe[c].unique()
```

```python
for (k,v) in class_labels:
    print('%s: %s' % (k,v))
```
